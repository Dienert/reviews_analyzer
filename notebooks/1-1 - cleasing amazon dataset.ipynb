```python
import pandas as pd
from datetime import datetime
import locale 
print("locale", "=>", locale.setlocale(locale.LC_ALL, "pt_BR.UTF-8"))
```

```python
df_amazon = pd.read_csv('../dataset/raw/amazon.csv')
df_amazon
```

```python
def get_only_date(raw_date):
    parts= raw_date.split(" em ")
    date = ""
    if len(parts) > 1:
        date = datetime.strptime(parts[1], '%d de %B de %Y').strftime('%Y-%m-%d')
    return date
```

```python
df_amazon['country'] = df_amazon.date.apply(lambda text: ' '.join(text.split(" em ")[0].split(" ")[2:]))
df_amazon['date'] = df_amazon.date.apply(get_only_date)

# joining title and review into review
df_amazon['review'] = df_amazon.title + ". " + df_amazon.review
```

```python
df_amazon = df_amazon[df_amazon.country.isin(['Estados Unidos', 'Reino Unido', 'Canadá', 'Austrália'])]
df_amazon
```
