```python
import pandas as pd
from datetime import datetime
import locale 
print("locale", "=>", locale.setlocale(locale.LC_ALL, "en.UTF-8"))
```

```python
df_goodreads = pd.read_csv('../dataset/raw/goodreads.csv')
df_goodreads
```

```python
def parse_date(raw_date):
    date = ''
    date = datetime.strptime(raw_date, '%b %d, %Y').strftime('%Y-%m-%d')
    return date
```

```python
df_goodreads['date'] = df_goodreads.date.apply(parse_date)
```
