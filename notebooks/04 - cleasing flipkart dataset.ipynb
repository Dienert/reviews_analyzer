```python
import pandas as pd
from datetime import datetime
import locale 
print("locale", "=>", locale.setlocale(locale.LC_ALL, "en.UTF-8"))
```

```python
df_flipkart = pd.read_csv('../dataset/raw/flipkart.csv')
df_flipkart
```

```python
def parse_date(raw_date):
    date = ''
    if "ago" not in raw_date: 
        date = datetime.strptime(raw_date, '%b, %Y').strftime("%Y-%m")
    return date
```

```python
df_flipkart['date'] = df_flipkart.date.apply(parse_date)

# joining title and review into review
df_flipkart['review'] = df_flipkart.title + ". " + df_flipkart.review
```

```python
df_flipkart.to_csv('../dataset/cleansed/flipkart.csv', index=False)
```
