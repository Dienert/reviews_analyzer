###  Split the texto into sentences and grouping all in one list called 'dataset'

```python
dataset = nltk.sent_tokenize(text_reviews)
```

```python
# list of sentences
dataset
```

```python
#preprocessing
for i in range (len(dataset)):
    dataset[i] = dataset[i].lower() #converting all words to lower case
print (dataset)
```

### Creating a histogram (a list that informs the number of times a word appears in the text)

```python
# creating a histogram

def createHistogram(_dataset):
    word2count = {}
    for data in _dataset:
        words = nltk.word_tokenize(data)
        for word in words:
            if word not in word2count.keys():
                word2count[word] = 1
            else:
                word2count[word] += 1
                
    return word2count
```

```python
word2count = createHistogram(dataset)
```

```python
word2count
```

### Sorting the list considering the terms that appear in the list more often

```python
import heapq

# sorting the list to check the most frequent words 
freq_words = heapq.nlargest(50,word2count, key=word2count.get)
print (freq_words)
```

### Creating the bag of words putting a boolean information for each document saying if each word in the bag of words is in the document or not  

```python
def createBagOfWords(_dataset):
    list = []

    for data in _dataset:
        vector = []
        for word in freq_words:
            if word in nltk.word_tokenize(data):
                vector.append(1) # 1, if word is in document 
            else:
                vector.append(0) # 0, if word is not in document 
        list.append(vector)
        
    bag_of_words = np.asarray(list)
    return bag_of_words
```

```python
bag = createBagOfWords(dataset)
print(bag)
```
